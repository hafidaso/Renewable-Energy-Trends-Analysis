# Analyze key trends: Production over time by Energy Type and Country

```python
from tabulate import tabulate


try:
    # Grouping and summing production by year and energy type
    production_by_year_energy = train_df.pivot_table(
        values='Production (GWh)',
        index='Year',
        columns='Energy Type',
        aggfunc='sum'
    )
    
    # Grouping and summing production by year and country
    production_by_year_country = train_df.pivot_table(
        values='Production (GWh)',
        index='Year',
        columns='Country',
        aggfunc='sum'
    )
    
    # Displaying the results
    print('Production (GWh) by Year and Energy Type:\n', production_by_year_energy.to_markdown())
    print('\nProduction (GWh) by Year and Country:\n', production_by_year_country.to_markdown())

except Exception as e:
    print(f'An error occurred: {e}')
```

### Production by Energy Type Over Time

Our analysis of **Production (GWh) by Year and Energy Type** reveals fluctuating but generally increasing trends across different renewable energy sources. While there isn't a consistent year-on-year increase for all types, the overall capacity and output of renewable energy have grown significantly over the years. Hydro and Wind energy types show substantial production, with Solar and Biomass also contributing significantly. Geothermal production also shows a steady contribution.

### Production by Country Over Time ### 

The **Production (GWh) by Year and Country** data indicates varying levels of renewable energy production across different nations. Some countries consistently show high production, while others have more sporadic contributions. This suggests diverse approaches to renewable energy adoption and development, influenced by factors such as natural resources, policy, and economic conditions.

```python
try:
    # Calculate investment efficiency
    train_df['Investment Efficiency'] = train_df['Production (GWh)'] / train_df['Investments (USD)']
    
    # Calculate average investment efficiency per country
    investment_efficiency_by_country = (
        train_df.groupby('Country')['Investment Efficiency']
        .mean()
        .sort_values(ascending=False)
    )
    
    print(
        'Average Investment Efficiency by Country (Production GWh per USD Investment):\n',
        investment_efficiency_by_country.to_markdown()
    )
    
    # Calculate correlation of numerical features with Production (GWh)
    numerical_cols = train_df.select_dtypes(include=['float64', 'int64']).columns
    numerical_cols = numerical_cols.drop(['ID', 'Year'], errors='ignore')
    
    production_correlation = train_df[numerical_cols].corr()['Production (GWh)'].sort_values(ascending=False)
    
    print('\nCorrelation of Features with Production (GWh):\n', production_correlation.to_markdown())

except Exception as e:
    print(f'An error occurred: {e}')
```

### Regions Investing Most Efficiently in Renewables

China, Japan, and India appear to be the most efficient in terms of GWh produced per USD invested, indicating that their investments are yielding higher energy outputs compared to other nations in the dataset. This could be attributed to factors such as lower labor costs, efficient project management, or specific policy incentives that maximize return on investment in renewable energy infrastructure.

**Positive Correlations:**

‚Ä¢ **Investment Efficiency, Average Annual Temperature, Public-Private Partnerships in Energy, Renewable Energy Targets, Solar Irradiance, Electricity Prices, Proportion of Energy from Renewables, Wind Speed, Hydro Potential, Innovation Index, and Energy Market Liberalization** show a positive correlation with **Production (GWh)**. This suggests that higher values in these factors are generally associated with increased renewable energy production. Notably, **Investment Efficiency** has the strongest positive correlation, reinforcing the importance of efficient investment strategies.



**Negative Correlations:**

‚Ä¢ **Factors such as CO2 Emissions, Energy Consumption, Renewable Energy Patents, Political Stability, Investments (USD), Population, Control of Corruption, Regulatory Quality, Ease of Doing Business, Biomass Availability, Energy Storage Capacity, Energy Subsidies, Grid Integration Capability, R&D Expenditure, Installed Capacity (MW), GDP, and Government Policies** show a negative correlation with Production (GWh). While some of these might seem counter-intuitive (e.g., Investments (USD) having a negative correlation), it's important to remember that correlation does not imply causation. Further in-depth analysis would be required to understand the underlying dynamics. For instance, a negative correlation with Investments (USD) might indicate that countries with already high production levels require less new investment to maintain or slightly increase output, or that initial large investments have already yielded their primary returns.



```python
try:
    # Calculate investment efficiency
    train_df['Investment Efficiency'] = train_df['Production (GWh)'] / train_df['Investments (USD)']
    
    # Average investment efficiency per country
    investment_efficiency_by_country = (train_df.groupby('Country')['Investment Efficiency'].mean().sort_values(ascending=False))
    
    # Plot investment efficiency by country
    plt.figure(figsize=(12, 7))
    sns.barplot(x=investment_efficiency_by_country.index, y=investment_efficiency_by_country.values, palette='viridis')
    plt.title('Average Investment Efficiency by Country (GWh/USD)')
    plt.xlabel('Country')
    plt.ylabel('Efficiency (GWh/USD)')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f'An error occurred: {e}')
```

```python
# Select only numeric variables
numeric_df = train_df.select_dtypes(include=[np.number])

# Correlation matrix
corr = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(14,10))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Define features and target
X = train_df.drop(columns=['Production (GWh)'])
y = train_df['Production (GWh)']

# One-hot encode categorical variables
X = pd.get_dummies(X, drop_first=True)

# Convert all columns to float (ensures numeric dtype)
X = X.astype(float)
y = y.astype(float)

# Add constant term
X = sm.add_constant(X)

# Fit model
model = sm.OLS(y, X).fit()

# summary
print(model.summary())
```

We applied Multiple Linear Regression (OLS) to better understand the multivariate influence on renewable energy production.

**Key Findings:**
R-squared: 0.017 ‚Äî the model explains only ~1.7% of the variance in Production (GWh), suggesting a weak linear fit.

**Significant Predictor:**

Year (p < 0.001): Positive and significant, confirming temporal growth in production.

Average Annual Temperature (p = 0.003): Positively associated with production, possibly due to solar potential.

Investment Efficiency (p ‚âà 0.068): Marginally significant, supporting its correlation-based importance.

Non-significant Factors: Most economic, policy, and population-based factors have high p-values (>0.1), indicating low individual significance in this linear model.

**Model Limitations:**
The condition number is high (3.43e+17), suggesting multicollinearity among predictors, which can distort coefficient estimates.

Despite low R¬≤, the model is useful for hypothesis testing and identifying non-linear or interaction patterns for further study.



```python
# Interaction between Investment and Political_Stability
train_df['Invest_PS'] = train_df['Investments (USD)'] * train_df['Political Stability']

X_inter = train_df[['Investments (USD)', 'Political Stability', 'Invest_PS']]
X_inter = sm.add_constant(X_inter)
y = train_df['Production (GWh)']

model_inter = sm.OLS(y, X_inter).fit()
print(model_inter.summary())
```

**Interactions and Multicollinearity Consideration**
An exploratory interaction term between Investments and Political Stability was tested but did not yield statistical significance (p = 0.323). However, the model's weak performance may also stem from:

* Highly correlated predictors
* Unobserved latent variables (e.g., technological readiness, workforce skills)
* Time-lag effects not captured by static regression
